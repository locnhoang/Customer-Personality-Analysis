```
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
# Read csv file & Show data set table
df = pd.read_csv("marketing_campaign.csv", sep ="\t")
df
```

```
# Assuming 'df' is your DataFrame containing the dataset
# Replace 'Age' with the actual variable name from your dataset
sorted_df = df.sort_values(by=['NumWebVisitsMonth','NumCatalogPurchases','NumWebPurchases'], ascending=False)

# Print the sorted DataFrame
sorted_df
```

```

#First 100 customers
df_subset = df.head(20)
# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df_subset['Dt_Customer'], df_subset['NumWebPurchases'], label='NumWebPurchases', alpha=0.5)


# Set labels and title
plt.xlabel('Date Joined')
plt.ylabel('Number of Web Purchases')
plt.title('Comparison of Variables to NumWebPurchases Over Time')
plt.legend()

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show plot
plt.tight_layout()
plt.show()
```

```
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])

# Group by year and sum the number of web purchases
grouped_df = df.groupby(df['Dt_Customer'].dt.year)['NumWebPurchases'].sum()

# Plot
plt.figure(figsize=(10, 6))
plt.bar(grouped_df.index, grouped_df.values, color='skyblue')

# Set labels and title
plt.xlabel('Year Joined')
plt.ylabel('Total Number of Web Purchases')
plt.title('Total Number of Web Purchases by Year Joined')

# Show plot
plt.tight_layout()
plt.show()
```

```

# Convert 'Dt_Customer' column to datetime type
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])

# Group by year and sum the number of purchases for each type
grouped_df = df.groupby(df['Dt_Customer'].dt.year)[['NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']].sum()

# Plotting
plt.figure(figsize=(10, 6))

# Bar width
bar_width = 0.2

# Position of bars on x-axis
r1 = range(len(grouped_df.index))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

# Plotting bars for each type of purchase
plt.bar(r1, grouped_df['NumWebPurchases'], color='skyblue', width=bar_width, label='NumWebPurchases')
plt.bar(r2, grouped_df['NumCatalogPurchases'], color='orange', width=bar_width, label='NumCatalogPurchases')
plt.bar(r3, grouped_df['NumStorePurchases'], color='green', width=bar_width, label='NumStorePurchases')

# Set labels and title
plt.xlabel('Year Joined')
plt.ylabel('Total Number of Purchases')
plt.title('Total Number of Purchases by Year Joined')
plt.xticks([r + bar_width for r in range(len(grouped_df.index))], grouped_df.index)

# Show legend
plt.legend()

# Show plot
plt.tight_layout()
plt.show()
```

```
import matplotlib.pyplot as plt
import pandas as pd

# Assuming 'df' is your DataFrame containing the dataset
# Replace 'Dt_Customer' with the actual column name containing the date they've been a member
# Replace 'NumWebPurchases', 'NumCatalogPurchases', and 'NumStorePurchases' with the actual column names

# Convert 'Dt_Customer' column to datetime type
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])

# Group by month and year and sum the number of purchases for each type
grouped_df = df.groupby([df['Dt_Customer'].dt.year, df['Dt_Customer'].dt.month])[
    ['NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']].sum()

# Convert index to datetime
grouped_df.index = pd.to_datetime(grouped_df.index.map(lambda x: '-'.join(map(str, x))))

# Plotting
plt.figure(figsize=(12, 6))

# Plotting lines for each type of purchase
plt.plot(grouped_df.index, grouped_df['NumWebPurchases'], label='NumWebPurchases', marker='o')
plt.plot(grouped_df.index, grouped_df['NumCatalogPurchases'], label='NumCatalogPurchases', marker='o')
plt.plot(grouped_df.index, grouped_df['NumStorePurchases'], label='NumStorePurchases', marker='o')

# Set labels and title
plt.xlabel('Month and Year Joined')
plt.ylabel('Total Number of Purchases')
plt.title('Total Number of Purchases Over Time by Month and Year Joined')

# Show legend
plt.legend()

# Show plot
plt.tight_layout()
plt.show()
```
